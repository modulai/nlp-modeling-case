```python
def get_accuracy(prediction, label):
    acc = (prediction.round() == label).mean()
    import pdb;pdb.set_trace()
    acc = float(acc)

    return acc
```

```python
from data import split_data

# Load the small dataset to see what it looks like
df_small = pd.read_csv("data/twitter_dataset_small_w_bart_preds.csv")
train_df, eval_df = split_data(df_small, percent_train=0.8)
get_accuracy(eval_df['bart_is_positive'].values, eval_df['is_positive'].values)
```

```python
# Quick look at the distribution of labels in train and eval datasets

no_samples_train = train_df.is_positive.shape[0]
no_pos_samples_train = train_df.is_positive.sum()
no_not_pos_samples_train = no_samples_train - no_pos_samples_train
print('Number of positive samples train: {}'.format(no_pos_samples_train))
print('Number of not positive samples train: {}'.format(no_not_pos_samples_train))

no_samples_eval = eval_df.is_positive.shape[0]
no_pos_samples_eval = eval_df.is_positive.sum()
no_not_pos_samples_eval = no_samples_eval - no_pos_samples_eval
print('Number of positive samples eval: {}'.format(no_pos_samples_eval))
print('Number of not positive samples eval: {}'.format(no_not_pos_samples_eval))
```
